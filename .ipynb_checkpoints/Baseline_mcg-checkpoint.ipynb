We'll work out of the 224x224 image set, but of course we can modify this however we want!

```python
path = Path('../data/flowers/tfrecords-jpeg-224x224')
```

```python
train_fnames = get_files(path/'train', extensions='.tfrec')
test_fnames = get_files(path/'val', extensions='.tfrec')
```

In order to parse the data, we need to "unpack" each sample to bring it into a `Dataset`. `tfrecord` has two special functions we can use, `unpack_int_64_list` and `unpack_bytes_list` (our classes are integers and our images are bytes)

```python
def unpack_sample(feats):
    return {
        'class' : unpack_int64_list(feats['class']),
        'image' : unpack_bytes_list(feats['image'])
    }
```

The current file situation (for this folder) is like so:

* 16 `tfrecord` files each with 792 images for training and testing

How would we split with that? We'll use an index splitter for everything after 12753

```python
data = []
for name in train_fnames+test_fnames:
    r = Reader(str(name), unpack_sample)
    for sample in r:
        data.append([sample['image'][0], sample['class'][0]])
```

```python
len(data)
```

Now we'll make our getters. They'll operate by grabbing the first and second items from a particular `x`

```python
get_x = lambda o: PILImage.create(io.BytesIO(o[0]))
get_y = lambda o: o[1]
```

```python
def get_items(noop): return data
```

```python
splitter = IndexSplitter(range(12753, len(data)))
```

```python
# data=data[:1000]
# def get_items(noop): return data
# splitter = IndexSplitter(range(500, len(data)))
```

```python
block = DataBlock(blocks=(ImageBlock, CategoryBlock), 
                  get_items=get_items,
                  get_x=get_x,
                  get_y=get_y,
                  splitter=splitter,
                  item_tfms=[Resize(224)],
                  batch_tfms=[*aug_transforms(mult=2)])
```

```python
dls = block.dataloaders('', bs=64)
len(dls.train), len(dls.valid), dls.c
```

```python
dls.show_batch()
```

Now we'll build a model and train!

```python
#cbs = [ContrastCallback(size=SZ, aug_targ=aug, aug_pos=aug2, temperature=0.3)]

m=partial(xresnet50, pretrained=False, sa=True, act_cls=Mish)

learn = cnn_learner(dls, m, opt_func=ranger, metrics=accuracy) #, cbs=cbs)

learn.unfreeze()
learn.summary()
```

Try again, with pretrained=False

```python
def do_train(runs=1,  num_ep=5, lr=1e-2):
    acc_res,val_res=[],[]
    for r in range(runs):
        print(f'Run: {r+1}')
        # GET LEARNER
        m=partial(xresnet50, pretrained=False, sa=True, act_cls=Mish)
        learn = cnn_learner(dls, m, opt_func=ranger, metrics=accuracy, pretrained=False).to_fp16() #, cbs=cbs)
        # RESET PARAMETERS (to be sure)
        #init_cnn(learn.model)
        # TRAIN
        learn.unfreeze()
        learn.fit_flat_cos(num_ep, lr)
        # LOG
        acc_res.append(learn.final_record[2])
        val_res.append(learn.final_record[1])
        #print(learn.final_record)
        #print(learn.recorder.log)
        # CLEAR MEM
        if r < runs-1:
            learn=None
            del learn
            torch.cuda.empty_cache()
            gc.collect()
            print('')

    print(f'Average Val Acc: {sum(acc_res)/len(acc_res):.2f}% {stdev(acc_res):.2f}%')
    return acc_res, val_res, learn
```

```python
acc_res, val_res, learn = do_train(runs=5, num_ep=5, lr=1e-2)
```

```python
learn.show_results()
```

Lets try again, with stronger augmentations, mult=2

```python
def do_train(runs=1,  num_ep=5, lr=1e-2):
    acc_res,val_res=[],[]
    for r in range(runs):
        print(f'Run: {r+1}')
        # GET LEARNER
        m=partial(xresnet50, pretrained=False, sa=True, act_cls=Mish)
        learn = cnn_learner(dls, m, opt_func=ranger, metrics=accuracy, pretrained=False).to_fp16() #, cbs=cbs)
        # TRAIN
        learn.unfreeze()
        learn.fit_flat_cos(num_ep, lr)
        # LOG
        acc_res.append(learn.final_record[2])
        val_res.append(learn.final_record[1])
        #print(learn.final_record)
        #print(learn.recorder.log)
        # CLEAR MEM
        if r < runs-1:
            learn=None
            del learn
            torch.cuda.empty_cache()
            gc.collect()
            print('')

    print(f'Average Val Acc: {(sum(acc_res)/len(acc_res))*100:.2f}% {stdev(acc_res):.2f}%')
    print(f'Average Val Acc: {(sum(val_res)/len(val_res))*100:.2f}% {stdev(val_res):.2f}%')
    return acc_res, val_res, learn
```

```python
acc_res, val_res, learn = do_train(runs=3, num_ep=5, lr=1e-2)
```
