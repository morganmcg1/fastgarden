```python
import tensorflow as tf
```

```python
# %load_ext autoreload
# %autoreload 2

from fastai2.vision.all import *
from statistics import stdev
from tfrecord.tfrecord import *
import gc
```

```python
path = Path('../data/flowers/tfrecords-jpeg-224x224')

train_fnames = get_files(path/'train', extensions='.tfrec')
test_fnames = get_files(path/'val', extensions='.tfrec')

def unpack_sample(feats):
    return {
        'class' : unpack_int64_list(feats['class']),
        'image' : unpack_bytes_list(feats['image'])
    }
```

```python
data = []
for name in train_fnames+test_fnames:
    r = Reader(str(name), unpack_sample)
    for sample in r:
        data.append([sample['image'][0], sample['class'][0]])
```

```python
len(data)
```

```python
get_x = lambda o: PILImage.create(io.BytesIO(o[0]))
get_y = lambda o: o[1]

def get_items(noop): return data

splitter = IndexSplitter(range(12753, len(data)))
```

```python
block = DataBlock(blocks=(ImageBlock, CategoryBlock), 
                  get_items=get_items,
                  get_x=get_x,
                  get_y=get_y,
                  splitter=splitter,
                  item_tfms=[Resize(224)],
                  batch_tfms=[*aug_transforms()])
```

```python
dls = block.dataloaders('', bs=64)
len(dls.train), len(dls.valid), dls.c
```

```python
dls.show_batch()
```

## 1. Baseline - 63%

```python
def do_train(runs=1,  num_ep=5, lr=1e-2):
    acc_res,val_res=[],[]
    for r in range(runs):
        print(f'Run: {r+1}')
        # GET LEARNER
        m=partial(xresnet50, pretrained=False, sa=True, act_cls=Mish)
        learn = cnn_learner(dls, m, opt_func=ranger, metrics=accuracy, pretrained=False).to_fp16() #, cbs=cbs)
        # RESET PARAMETERS (to be sure)
        #init_cnn(learn.model)
        # TRAIN
        learn.unfreeze()
        learn.fit_flat_cos(num_ep, lr)
        # LOG
        acc_res.append(learn.final_record[2])
        val_res.append(learn.final_record[1])
        #print(learn.final_record)
        #print(learn.recorder.log)
        # CLEAR MEM
        if r < runs-1:
            learn=None
            del learn
            torch.cuda.empty_cache()
            gc.collect()
            print('')

    print(f'Average Val Acc: {sum(acc_res)/len(acc_res):.2f}% ±{stdev(acc_res):.2f}%')
    return acc_res, val_res, learn
```

```python
acc_res, val_res, learn = do_train(runs=5, num_ep=5, lr=1e-2)
```

```python
learn.recorder.plot_loss()
```

```python
learn.recorder.plot_sched()
```

## 10. Progressive Resizing (1e @128, 4e @ 224) - 68.75%

```python
def get_dls(sz=224):
    block = DataBlock(blocks=(ImageBlock, CategoryBlock), 
                      get_items=get_items,
                      get_x=get_x,
                      get_y=get_y,
                      splitter=splitter,
                      item_tfms=[Resize(sz)],
                      batch_tfms=[*aug_transforms()])
    dls = block.dataloaders('', bs=64)
    return dls
```

```python
def do_train(runs=1,  num_ep=5, lr=1e-2):
    acc_res,val_res=[],[]
    stg_1_num_ep = 1
    cbs=[] #[ColorfulProgressCallback()]
    for r in range(runs):
        print(f'Run: {r+1}')
        # GET LEARNER
        m=xresnet50(pretrained=False, sa=True, act_cls=Mish)
        loss_func=LabelSmoothingCrossEntropy()
        
        # STG 1
        dls = get_dls(sz=128)
        learn = Learner(dls, m, opt_func=ranger, loss_func=loss_func, metrics=accuracy).to_fp16()
        # TRAIN
        learn.unfreeze()
        learn.fit_flat_cos(stg_1_num_ep, lr, cbs=cbs, pct_start=0.9)
        
        # STG 2
        dls = get_dls(sz=224)
        learn.dls = dls
        learn.fit_flat_cos(num_ep-stg_1_num_ep, lr, cbs=cbs, pct_start=0.9)
        
        # LOG
        acc_res.append(learn.final_record[2])
        val_res.append(learn.final_record[1])
        # CLEAR MEM
        if r < runs-1:
            learn=None
            del learn
            torch.cuda.empty_cache()
            gc.collect()
            print('')

    print(f'Average Val Acc: {(sum(acc_res)/len(acc_res))*100:.2f}% ±{(stdev(acc_res)*100):.2f}%')
    print(f'Average Val Loss: {sum(val_res)/len(val_res):.2f} ±{stdev(val_res):.2f}')
    return acc_res, val_res, learn
```

```python
acc_res, val_res, learn = do_train(runs=3, num_ep=5, lr=1e-2)
```

```python
def do_train(runs=1,  num_ep=5, lr=1e-2):
    acc_res,val_res=[],[]
    stg_1_num_ep = 2
    cbs=[] #[ColorfulProgressCallback()]
    for r in range(runs):
        print(f'Run: {r+1}')
        # GET LEARNER
        m=xresnet50(pretrained=False, sa=True, act_cls=Mish)
        loss_func=LabelSmoothingCrossEntropy()
        
        # STG 1
        dls = get_dls(sz=128)
        learn = Learner(dls, m, opt_func=ranger, loss_func=loss_func, metrics=accuracy).to_fp16()
        # TRAIN
        learn.unfreeze()
        #learn.fit(stg_1_num_ep, lr)
        learn.fit_flat_cos(stg_1_num_ep, lr, cbs=cbs)
        
        # STG 2
        dls = get_dls(sz=224)
        learn.dls = dls
        learn.fit_flat_cos(num_ep-stg_1_num_ep, lr, cbs=cbs, pct_start=0.25)
        
        # LOG
        acc_res.append(learn.final_record[2])
        val_res.append(learn.final_record[1])
        # CLEAR MEM
        if r < runs-1:
            learn=None
            del learn
            torch.cuda.empty_cache()
            gc.collect()
            print('')

    print(f'Average Val Acc: {(sum(acc_res)/len(acc_res))*100:.2f}% ±{(stdev(acc_res)*100):.2f}%')
    print(f'Average Val Loss: {sum(val_res)/len(val_res):.2f} ±{stdev(val_res):.2f}')
    return acc_res, val_res, learn
```

```python
acc_res, val_res, learn = do_train(runs=3, num_ep=5, lr=1e-2)
```
